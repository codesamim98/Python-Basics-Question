```python
#2 Write a python program to read the contents of a file and print each line.

# File name to read
file_name = "example.txt"

try:
    # Open the file in read mode
    with open(file_name, "r") as file:
        # Read and print each line
        for line in file:
            print(line.strip())  # .strip() removes extra newline characters
except FileNotFoundError:
    print("Error: The file does not exist.")

```

```python
# 4. Write a python script that reads from one file and writes its content to another file.

try:
    # Open the source file for reading
    source = open("source.txt", "r")
    data = source.read()
    source.close()

    # Open the destination file for writing
    destination = open("destination.txt", "w")
    destination.write(data)
    destination.close()

    print("File copied successfully!")

except FileNotFoundError:
    print("The source file was not found.")

```

```python
 # 5. How would you catch and handle division by zero error in python?

 try:
    a = 10
    b = 0
    result = a / b  # This will cause a ZeroDivisionError
    print("Result:", result)

except ZeroDivisionError:
    print("You can't divide by zero!")
```

```python
#6 Write a python program that logs an error message to a log file when division by zero exception occurs.

import logging

# Set up logging to write to a file
logging.basicConfig(filename='error_log.txt', level=logging.ERROR)

try:
    a = 10
    b = 0
    result = a / b

except ZeroDivisionError:
    logging.error("Tried to divide by zero.")
    print("An error occurred. Check 'error_log.txt' for details.")
```

```python
 # 7. How do you log information at different levels(INFO,ERROR,WARNING) in python using the logging module ?

 import logging

# Set up logging configuration
logging.basicConfig(
    filename='app.log',
    level=logging.DEBUG,  # Capture all levels from DEBUG and above
    format='%(levelname)s - %(message)s'
)

# Logging messages
logging.debug("This is a DEBUG message.")
logging.info("This is an INFO message.")
logging.warning("This is a WARNING message.")
logging.error("This is an ERROR message.")
logging.critical("This is a CRITICAL message.")

print("Logs have been written to 'app.log'")
```

```python
#21 Write a python program that reads a file and prints the number ofoccurences of a specific word.

# Ask the user for the file name and the word to search
file_name = input("Enter the file name: ")
search_word = input("Enter the word to count: ")

try:
    with open(file_name, 'r') as file:
        content = file.read()
        # Convert content and search word to lowercase for case-insensitive matching
        word_count = content.lower().split().count(search_word.lower())
        print(f"The word '{search_word}' appears {word_count} times.")
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("An error occurred:", e)

```

```python
#22 How can you check if a file is empty before attempting to read its contents ?
import os

file_name = "example.txt"

if os.path.exists(file_name):
    if os.path.getsize(file_name) == 0:
        print("The file is empty.")
    else:
        with open(file_name, "r") as file:
            content = file.read()
            print("File content:")
            print(content)
else:
    print("The file does not exist.")

```
